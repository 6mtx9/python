```python
import numpy as np
x = np.linspace(0,5,10)
y = x**2
print(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,c='red',lw=3,ls='--') # c couleur lw largeur ligne ls style ligne
plt.show()
```

```python
plt.figure()
plt.plot(x,y,label='petit')
plt.plot(x,x**3,label='gros')
plt.title('figure')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
#plt.savefig('test.png')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,c='red')
plt.subplot(2,1,2)
plt.plot(x,y,c='blue')
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('graphique 1')
plt.subplot(2,2,2)
plt.plot(x,np.sin(x),label='sinus')
plt.plot(x,np.cos(x),label='cosinus')
plt.title('graphique 2')
plt.legend()
plt.show()
```

```python
fig, axe = plt.subplots(2,1,sharex=True)
axe[0].plot(x,y)
axe[1].plot(x,np.sin(x))
plt.show()
```

```python
dataset = {f'experience {i}':np.random.randn(100,3) for i in range (6)}
#dataset
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,15))
    # k est la localisation de la DATA -->> f'experience{i}'
    # i est la liste de n
    # n nombre de lignes de graphiques
    # 1 nombre de colonne de graphique ||||
    for k,i in zip(data.keys(),range(1,n+1)):
        plt.subplot(n,3,i) #
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
 graphique(dataset)
```
