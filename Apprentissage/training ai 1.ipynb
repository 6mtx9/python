```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data # dataset a classer
y = iris.target # objectifs de classification
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables') # shape[0]= nombres data test shape[1]= variables qui permettent de classer
print(f'il y a {np.unique(y).size} classes')
#names
```

```python
n = x.shape[1]
plt.figure(figsize=(12, 8))
for i in range(n):
    plt.subplot(n//2, n//2, i+1)
    plt.scatter(x[:, 0], x[:, i], c=y)
    plt.xlabel('0')
    plt.ylabel(i)
    plt.colorbar(ticks=list(np.unique(y)))
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1], c=y , alpha=0.5, s=x[:,2]*100) # c = y type de fleurs ( 3 types deja définis)  # alpha controle la transparence s = taille points
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
```

```python
# Graphique 3D
# lancer la demo direct sur jupiter en 3D
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2], c=y)
```

```python
f = lambda x, y: np.sin(x)+np.cos(x+y) # creation de fonction f(x,y): np.sin(x)+np.cos(x+y)

X = np.linspace(0,5,100) # creation variable X et Y de 100 points de 0 à 5
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)
Z = f(X,Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='plasma')
```

```python
# Histogramme

plt.hist(x[:,0], bins=20) # bins = selection dans histogramme
plt.hist(x[:,1], bins=20)
```

```python
plt.hist2d(x[:,0],x[:,1],cmap='Blues')
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.colorbar()
```

```python
from scipy import misc
face = misc.face(gray=True)
#plt.imshow(face, cmap='gray')
plt.hist(face.ravel(),bins=255) # ravel applatir sur une ligne tout
plt.xlabel('couleur pixel')
plt.ylabel('nombres apparitions')
plt.show()
```
