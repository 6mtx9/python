```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(0)
m = 100 # creation de 100 échantillons
X = np.linspace(0, 10, m).reshape(m,1)
y = X + np.random.randn(m, 1)

plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
model.score(X,y)

prediction = model.predict(X)
plt.scatter(X,y)
plt.plot(X,prediction,c='g')
```

```python
from sklearn.svm import SVR

np.random.seed(0)
m = 100 # creation de 100 échantillons
X = np.linspace(0, 10, m).reshape(m,1)
y = X**2 + np.random.randn(m, 1)

plt.scatter(X, y)
```

```python
model = SVR(C=100)
model.fit(X,y)
model.score(X,y)

prediction = model.predict(X)

plt.scatter(X,y)
plt.plot(X,prediction,c='g')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic = titanic[['survived', 'pclass', 'sex', 'age']]
titanic.dropna(axis=0, inplace=True)
titanic['sex'].replace(['male', 'female'], [0, 1], inplace=True)
titanic.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
y = titanic['survived']
X = titanic.drop('survived',axis=1)
X = X.values
```

```python
model.fit(X,y)
model.score(X,y)
```

```python
model.predict(X)
```

```python
def survie(model,pclass=3,sex=0,age=20):
    x = np.array([pclass,sex,age]).reshape(1,3)
    print(model.predict(x))
    print(model.predict_proba(x))
```

```python
survie(model)
```

```python
modeln = KNeighborsClassifier(n_neighbors=1)
```

```python
modeln.fit(X,y)
modeln.score(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
#random_state pour avoir aléatoire fixe
print('Train set:', X_train.shape)
print('Test set:', X_test.shape)
```

```python
plt.figure(figsize=(4,1))
plt.subplot(121)
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,alpha=0.8)
plt.title('Train set')
plt.subplot(122)
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,alpha=0.8)
plt.title('Test set')
```

```python
model = KNeighborsClassifier(n_neighbors=6)
model.fit(X_train,y_train)
print('Train score : ',model.score(X_train,y_train))
print('Test score : ',model.score(X_test,y_test))
# Comparaison entre train et la data test
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(KNeighborsClassifier(), X_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
val_score=[]
for i in range(1,50):
    score = cross_val_score(KNeighborsClassifier(i), X_train, y_train, cv=5, scoring='accuracy').mean()
    val_score.append(score)
    
plt.plot(val_score)
```

```python
from sklearn.model_selection import validation_curve
```

```python
model = KNeighborsClassifier()
k = np.arange(1, 50)
train_score, val_score = validation_curve(model, X_train, y_train, param_name='n_neighbors',param_range=k, cv=5)

plt.plot(k, val_score.mean(axis=1), label='validation')
plt.plot(k, train_score.mean(axis=1), label='train')

plt.ylabel('score')
plt.xlabel('n_neighbors')
plt.legend()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors': np.arange(1, 20),
              'metric': ['euclidean', 'manhattan']}

grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)

grid.best_score_ # search best score
```

```python
grid.best_params_
```

```python
model = grid.best_estimator_
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,model.predict(X_test))
```

```python
from sklearn.model_selection import learning_curve
```

```python
N, train_score, val_score = learning_curve(model, X_train, y_train,train_sizes=np.linspace(0.1, 1, 70), cv=5)

plt.plot(N, train_score.mean(axis=1), label='train')
plt.plot(N, val_score.mean(axis=1), label='validation')
plt.xlabel('train_sizes')
plt.legend()
```
