# Relation Variables / Target 

# Création de sous-ensembles positifs et négatifs

```python
data_positive = data[data['SARS-Cov-2 exam result'] == 'positive']
```

```python
data_negative = data[data['SARS-Cov-2 exam result'] == 'negative']
```

# Création des ensembles Blood et viral

```python
missing_rate = data.isna().sum()/data.shape[0]
```

```python
(data.isna().sum()/data.shape[0]).sort_values(ascending=True)
```

```python
blood_columns = data.columns[(missing_rate < 0.9)&(missing_rate > 0.88)]
```

```python
viral_columns = data.columns[(missing_rate < 0.88)&(missing_rate > 0.75)]
```

# Target / Blood

```python
for columns in blood_columns:
    plt.figure()
    sns.distplot(data_positive[columns],label='positive')
    sns.distplot(data_negative[columns],label='negative')
    plt.legend()
```

# Relation Target / age

```python
sns.countplot(x='Patient age quantile',hue='SARS-Cov-2 exam result',data=data)
```

# Relation Target / Viral

```python
pd.crosstab(data['SARS-Cov-2 exam result'], data['Influenza A'])
```

```python
for columns in viral_columns:
    plt.figure()
    sns.heatmap(pd.crosstab(data['SARS-Cov-2 exam result'],data[columns]), annot=True, fmt='d') # annot = annotation / fmt -> afficher en entier les nombres
```
